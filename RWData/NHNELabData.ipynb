```python
threshold = 2000;
# Calculate the number of missing values in each column
missing_counts = dataset.isnull().sum()

# Identify columns exceeding the threshold
cols_to_drop = missing_counts[missing_counts > threshold].index

# Drop the identified columns
dataset.drop(cols_to_drop, axis=1, inplace=True)

dataset
```

```python
dataset = dataset.dropna();
display(dataset);
# Randomly sample 70% of your dataframe
dataset = dataset.sample(frac=0.3);
display(dataset);
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2);

dataset.loc[:,:] = imputer.fit_transform(dataset.loc[:,:]);

dataset
```
