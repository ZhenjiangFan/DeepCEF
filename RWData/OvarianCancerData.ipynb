```python
data_path = "OvarianCancer/";
file_name = "OvarianCancer.csv";
dataset = pd.read_csv(data_path+file_name,index_col=0);
display(dataset);
#print(mixed_sim_dataset.columns);

# # dataset.replace({"NO": 0, "YES": 1}, inplace=True);
# print(dataset["Gravidity"].unique());
# dataset.replace({'1st':1, '2nd':2, '5th':5, '3rd':3, '4th':4, '6th':6, '7th':7}, inplace=True);
# print(dataset["Gender"].unique());
# dataset.replace({"Girl": 1, "girl": 1,"boy":2}, inplace=True);
# print(dataset["Baby-Weight(g)"].unique());
# dataset.replace({"A": 1, "B": 2,"C":3}, inplace=True);

# display(dataset);
```

```python
# dataset.isnull().values.any()
# Remove rows with any NaN values
# dataset.dropna(inplace=True);
# dataset.columns
# dataset['TYPE']

# Remove tab characters from all columns
dataset = dataset.replace('\t', '', regex=True);

dataset = dataset.replace('>', '', regex=True);

dataset = dataset.replace('<', '', regex=True);

display(dataset);
```

```python
# dataset['TYPE'].unique()
# Create an imputer instance
imputer = SimpleImputer(strategy='mean');

# Define the condition for imputation
condition = dataset['TYPE'] == 1;

# # Apply imputation only to rows that meet the condition
dataset.loc[condition, :] = imputer.fit_transform(dataset.loc[condition, :]);

# Define the condition for imputation
condition = dataset['TYPE'] == 0;

# # Apply imputation only to rows that meet the condition
dataset.loc[condition, :] = imputer.fit_transform(dataset.loc[condition, :]);

display(dataset);
```

```python
file_name = "processed_data.csv";
dataset.to_csv(data_path+file_name,index=None)
```
