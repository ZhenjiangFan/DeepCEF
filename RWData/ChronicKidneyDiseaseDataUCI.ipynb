```python
from ucimlrepo import fetch_ucirepo
import numpy as np
import pandas as pd

import sys
sys.path.append('../')

import Utils as Utils

  
# fetch dataset 
chronic_kidney_disease = fetch_ucirepo(id=336) 
  
# data (as pandas dataframes) 
X = chronic_kidney_disease.data.features 
y = chronic_kidney_disease.data.targets 
  
# metadata 
print(chronic_kidney_disease.metadata) 
  
# variable information 
print(chronic_kidney_disease.variables) 
```

```python
frames = [X, y];
dataset = pd.concat(frames, axis=1);
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
from pandas.api.types import is_string_dtype, is_numeric_dtype
from sklearn.utils.multiclass import type_of_target

dataset = Utils.string_to_numeric(dataset);
display(dataset);
```

```python
dataset['class'].unique()
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer instance
imputer = SimpleImputer(strategy='mean');

# Define the condition for imputation
condition = dataset['class'] == 1;

# # Apply imputation only to rows that meet the condition
dataset.loc[condition, :] = imputer.fit_transform(dataset.loc[condition, :]);

# Define the condition for imputation
condition = dataset['class'] == 2;

# # Apply imputation only to rows that meet the condition
dataset.loc[condition, :] = imputer.fit_transform(dataset.loc[condition, :]);

# Define the condition for imputation
condition = dataset['class'] == 3;

# # Apply imputation only to rows that meet the condition
dataset.loc[condition, :] = imputer.fit_transform(dataset.loc[condition, :]);

display(dataset);
```

```python
#Save the dataset
dataset.to_csv('ChronicKidneyDisease/processed_data.csv',index=None);
```
