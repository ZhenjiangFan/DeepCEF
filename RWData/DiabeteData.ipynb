```python
import numpy as np
import pandas as pd
import networkx as nx

# import Utils as Utils

import scipy.io as sio
from scipy.io import arff
import pandas as pd
```

```python
data_path = "Diabetes/";
file_name = "dataset_37_diabetes.arff";

data = arff.loadarff(data_path+file_name)
dataset = pd.DataFrame(data[0])
dataset.head()

print(dataset["class"].unique());
# Convert bytes to strings
dataset['class'] = dataset['class'].str.decode('utf-8')

dataset.replace({"tested_negative": 0, "tested_positive": 1}, inplace=True);
dataset
```

```python
file_name = "processed_data.csv";
dataset.to_csv(data_path+file_name,index=None);
```
