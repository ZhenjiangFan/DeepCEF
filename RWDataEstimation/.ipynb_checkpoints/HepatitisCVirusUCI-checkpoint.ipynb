### https://archive.ics.uci.edu/dataset/503/hepatitis+c+virus+hcv+for+egyptian+patients

```python
import numpy as np
import pandas as pd

from ucimlrepo import fetch_ucirepo
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
hepatitis_c_virus_hcv_for_egyptian_patients = fetch_ucirepo(id=503) 
  
# data (as pandas dataframes) 
X = hepatitis_c_virus_hcv_for_egyptian_patients.data.features 
y = hepatitis_c_virus_hcv_for_egyptian_patients.data.targets 
  
# metadata 
print(hepatitis_c_virus_hcv_for_egyptian_patients.metadata) 
  
# variable information 
print(hepatitis_c_virus_hcv_for_egyptian_patients.variables) 
```

```python
frames = [X, y];
dataset = pd.concat(frames, axis=1);
dataset
```

```python
dataset.drop(['Gender'], axis=1, inplace=True);
dataset.drop(['Fever'], axis=1, inplace=True);
dataset.drop(['Nausea/Vomting'], axis=1, inplace=True);
dataset.drop(['Headache '], axis=1, inplace=True);
dataset.drop(['Diarrhea '], axis=1, inplace=True);
# dataset.drop(['Fatigue & generalized bone ache '], axis=1, inplace=True);
dataset.drop(['Jaundice '], axis=1, inplace=True);
dataset.drop(['Epigastric pain '], axis=1, inplace=True);
```

```python
#Save the dataset
dataset.to_csv('HepatitisCVirusUCI/processed_data.csv',index=None);
```
