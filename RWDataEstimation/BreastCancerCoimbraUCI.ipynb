### https://archive.ics.uci.edu/dataset/451/breast+cancer+coimbra

```python
import numpy as np
import pandas as pd

from ucimlrepo import fetch_ucirepo
```

```python
# fetch dataset 
breast_cancer_coimbra = fetch_ucirepo(id=451) 
  
# data (as pandas dataframes) 
X = breast_cancer_coimbra.data.features 
y = breast_cancer_coimbra.data.targets 
  
# metadata 
print(breast_cancer_coimbra.metadata) 
  
# variable information 
print(breast_cancer_coimbra.variables) 
```

```python
frames = [X, y];
dataset = pd.concat(frames, axis=1);
# dataset = dataset.astype(float)
dataset
```

```python
#Save the dataset
dataset.to_csv('BreastCancerCoimbraUCI/processed_data.csv',index=None);
```
