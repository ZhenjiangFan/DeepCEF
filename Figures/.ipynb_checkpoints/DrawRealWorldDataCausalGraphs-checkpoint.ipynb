```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import networkx as nx
from pyvis.network import Network

import pygraphviz as pgv

# import sys,os
# sys.path.insert(0,os.path.realpath('..'))
```

```python
def convert_networkx_to_pyvis(directed_graph):
    # g.from_nx(directed_graph);
    #,select_menu=True,filter_menu=True 
    g = Network(notebook=True,cdn_resources='in_line',directed=True);

    node_list = directed_graph.nodes();
    for node in node_list:
        g.add_node(node,shape="circle");
        
    for edge in directed_graph.edges():
        #print(edge[0]);
        g.add_edge(edge[0],edge[1]);
    return g;
```

```python
data_name_list = ["ChronicKidneyDisease","Diabetes","national-health-and-nutrition-examination-survey","OvarianCancer"
                 ,"RNA-Seq-Long-Read Sequencing of an Advanced Cancer Cohort Resolves Rearrangements, Unravels Haplotypes, and Reveals Methylation Landscapes"
                 ,"scRNA-seq-A Molecular Switch from Tumor Suppressor to Oncogene in ER-Positive Breast Cancer - Role of Androgen Receptor, JAK-STAT, and Lineage Plasticity"
                 ,"scRNA-seq-Gene expression profile at single cell level of a rare seen and young tracheal adenoid cystic carcinoma"
                 ,"scRNA-Seq-Spatial Multiomics Reveals Metabolic Reprogramming and Calcification Characteristics of Diabetic Macroangiopathy"
                 ,"Sepsis","SmokerStatusPredictionBio-Signals"];



for data_name in data_name_list:
    output_file = "ges_fci_pc_output_file.csv";
    # data_name = "ChronicKidneyDisease";
    file_path = "../RWData/"+data_name+"/";

    causal_relations = pd.read_csv(file_path+output_file);
    display(causal_relations.head());
    
    # var_info_file = "Variable name and descriptions.csv";
    # var_info = pd.read_csv(file_path+var_info_file);
    # display(var_info.head());
    
    directed_graph = nx.DiGraph();
    
    
    for idx,row in causal_relations.iterrows():
        cause = row["Cause"];
        effect = row["Effect"];
        
        # #if cause in var_info["Variable Name"].values:
        # cause_description = var_info.loc[var_info["Variable Name"]==cause]["Description"].values[0];
        # print(description);
        # effect_description = var_info.loc[var_info["Variable Name"]==effect]["Description"].values[0];
        # print(effect_description);
        
        directed_graph.add_edge(cause,effect);
    
        # g.prep_notebook()
        pyvis_G = convert_networkx_to_pyvis(directed_graph)
        pyvis_G.show("ValidatedRawDataCausalGraphs/"+data_name+"_causal_graph.html");
    
        # Convert the NetworkX graph to a PyGraphviz AGraph object
        PyGraphviz_G = nx.nx_agraph.to_agraph(directed_graph)
        PyGraphviz_G.layout("dot")  # layout with dot
        PyGraphviz_G.draw("ValidatedRawDataCausalGraphs/"+data_name+"_causal_graph.png")  # write to file
```
